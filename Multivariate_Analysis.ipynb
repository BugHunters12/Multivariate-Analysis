```python
###Insert file path/name here. You might need to upload the file each time openning. Link to data: https://hankenfi.sharepoint.com/:x:/t/Quantum1/EVfo4rFqfUtOpQRe7-pHq3IBNQJEMLGWlkIk3cwEqfmWiQ?e=Er86R4
dt = pd.read_excel('data_quantum.xlsx')
dt.head()
```

```python
### Rename column and remove unncessary columns using dictionaries. Run only once 

data = dt.rename(columns = {'Unnamed: 0': 'Month'}, inplace = False)

del(data['Unnamed: 26'])
del(data['Unnamed: 27'])
del(data['Unnamed: 53'])
del(data['Unnamed: 59'])


```

```python
data.head()
```

```python
data.info()
```

```python
###One Factor regression with a constant
y = data['Portfolio 1 excess return']
x=data['Factor 3']
x = sm.add_constant(x)
model = sm.OLS(y,x)

#Print the regression coefficients
results = model.fit()
results.params
results.tvalues

print(results.summary())
```

```python
###MultiFactor regression with a constant

y = data['Portfolio 1 excess return']
X=data[['Factor 1','Factor 2','Factor 3','Factor 4','Factor 5']]               
X = sm.add_constant(X)
model = sm.OLS(y,X)
results = model.fit()

#Print the regression coefficients
results.params
results.tvalues
R2 = print(results.rsquared)

print(results.summary())
```

```python
#Looping accross 25 portfolio excess return and run 25 regressions on 5 factors
    
for i in range(1,26):
    y= (data.iloc[:,i])
    X=data[['Factor 1','Factor 2','Factor 3','Factor 4','Factor 5']]
    X = sm.add_constant(X)
    model = sm.OLS(y,X)
    results = model.fit()
    
    results.params
    results.tvalues
    
    print(results.summary())
          
```
